```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the decentralized evolution file
base_dir = os.path.dirname(os.path.abspath('compare_fitness.ipynb'))
decentralized_path = os.path.join(base_dir, 'Individual Legs Parallel', 'controllers', 'indiv_legs', 'data', 'evolution_data.csv')
decentralized_data = pd.read_csv(decentralized_path)

# Filter data to include only generations up to 1800
decentralized_data = decentralized_data[decentralized_data['Generation'] <= 1800]

# Get unique leg indices
leg_indices = decentralized_data['Leg Index'].unique()

# Create a plot for each leg on the same graph
plt.figure(figsize=(30, 10))

leg_index = 0
# Filter data for leg 0
leg_data = decentralized_data[decentralized_data['Leg Index'] == leg_index]

# Group by generation and calculate the mean fitness for each generation
fitness_over_time = leg_data.groupby('Generation')['Fitness'].mean()

# Plot the fitness over time for the current leg
plt.plot(fitness_over_time, label='Decentralized Mean Fitness', color='#FFA800')

# Load the centralized evolution file
centralized_path = os.path.join(base_dir, 'Original', 'controllers', 'leg_3_disabled_original_ga', 'data', 'evolution_data.csv')
centralized_evolution = pd.read_csv(centralized_path)

# Filter data to include only generations up to 1800
centralized_data = centralized_evolution[centralized_evolution['Generation'] <= 1800]

# Calculate mean fitness for each generation
mean_fitness = centralized_data.groupby('Generation')['Fitness'].mean()

# Plot the data
plt.plot(mean_fitness.index, mean_fitness.values, linestyle='-', color='b', label='Centralized Mean Fitness')

# Add title and labels
plt.title('Mean Fitness Over Time for Each Leg - Decentralized vs Centralized Evolution - Leg 3 Disabled')
plt.xlabel('Generation')
plt.ylabel('Mean Fitness')
plt.grid(True)

# Set y/x-axis limits
plt.ylim(0, 90)
plt.xlim(0, 1800)

# Add legend outside the plot
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

# Adjust layout to make room for the legend
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Show plot
plt.show()
```
