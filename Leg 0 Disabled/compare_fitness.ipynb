```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the central evolution file
base_dir = os.path.dirname(os.path.abspath('compare_fitness.ipynb'))
centralized_path = os.path.join(base_dir, 'Original', 'controllers', 'leg_0_disabled_original_ga', 'data', 'evolution_data.csv')
centralized_evolution = pd.read_csv(centralized_path)

# Extract relevant columns
generations = centralized_evolution['Generation']
fitness = centralized_evolution['Fitness']

# Filter data to include only generations up to 1800
centralized_data = centralized_evolution[centralized_evolution['Generation'] <= 1800]

# Calculate mean fitness for each generation
mean_fitness = centralized_data.groupby('Generation')['Fitness'].mean()

# Plot the data
plt.figure(figsize=(30, 10))
plt.plot(mean_fitness.index, mean_fitness.values, linestyle='-', color='b', label='Mean Fitness')

# Add labels and title
plt.xlabel('Generation')
plt.ylabel('Mean Fitness')
plt.title('Mean Fitness Over Time for Each Leg - Centralized Evolution - Leg 0 Disabled')
plt.grid(True)

# Set y/x-axis limits
plt.ylim(0, 90)
plt.xlim(0, 1800)

# Add legend outside the plot
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

# Adjust layout to make room for the legend
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Show the plot
plt.show()
```
